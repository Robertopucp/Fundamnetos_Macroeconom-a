```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import xlrd
import os
from openpyxl import load_workbook
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
base = pd.read_excel(r'..\datos\Tipo_cambio.xlsx', sheet_name = "Big_MAC_2")
base = base.set_index('periodo')
```

```python

fig, ax = plt.subplots(figsize=(12,5))

ax = sns.lineplot( data = base)
plt.legend('')
plt.ylabel('Tasa de política monetaria')
plt.xlabel(' ')
plt.title('Devaluación de la moneda local')

ax.figure.legend(loc='lower center', bbox_to_anchor=(1.1, 0.5), ncol=1)
```

```python
sns.set('notebook')

ax = sns.lineplot(x= "periodo" , y= "Peru", data=base, 
                        label = 'Perú ', color="k", linestyle='--', legend=False)

#sns.despine()
plt.ylabel(' ')
plt.xlabel(' ')
plt.title('Devaluación de la moneda local - países de Latinoamérica');

ax2 = sns.lineplot(x= "periodo", y= "Argentina", data=base, ax=ax, color = "red", 
                         label ='Argentina', legend=False) 
ax3 = sns.lineplot(x= "periodo", y= "Brazil", data=base, ax=ax, color = "green", 
                         label ='Brazil', legend=False) 
ax4 = sns.lineplot(x= "periodo", y= "Chile", data=base, ax=ax, color = "blue", 
                         label ='Chile', legend=False) 

ax4 = sns.lineplot(x= "periodo", y= "Mexico", data=base, ax=ax, color = "brown", 
                         label ='Mexico', legend=False) 




ax.figure.legend(loc='lower center', bbox_to_anchor=(1.1, 0.5), ncol=1);


txt="Elboración propia: Fuente: The Economist"  
plt.figtext(0.5, 0.01, txt, wrap=True, horizontalalignment='right', fontsize=10)


```
